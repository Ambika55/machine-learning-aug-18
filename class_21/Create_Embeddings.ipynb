```python
import numpy as np
import os
from keras.applications.resnet50 import ResNet50
from keras.preprocessing import image
from keras.optimizers import Adam
from keras.layers import Dense,Activation,Flatten,Input,Dropout,GlobalAveragePooling2D
from keras.models import Model
from keras.utils import np_utils
import random

import matplotlib.pyplot as plt
```

```python
images_dir = os.listdir("./Flicker8k_Dataset/")

images_path = './Flicker8k_Dataset/'
captions_path = './Flickr8k_text/Flickr8k.token.txt'
train_path = './Flickr8k_text/Flickr_8k.trainImages.txt'
val_path = './Flickr8k_text/Flickr_8k.devImages.txt'
test_path = './Flickr8k_text/Flickr_8k.testImages.txt'

captions = open(captions_path, 'r').read().split("\n")
x_train = open(train_path, 'r').read().split("\n")
x_val = open(val_path, 'r').read().split("\n")
x_test = x_val = open(test_path, 'r').read().split("\n")
```

```python
train_data = []
ctr=0
for ix in x_train:
    if ix == "":
        continue
    ctr+=1
    if ctr%100==0:
        print(ctr)
    path = images_path + ix
    img = image.load_img(path,target_size=(224,224))
    img_array = image.img_to_array(img)
    train_data.append(img_array)
train_data = np.array(train_data)
```

```python
test_data = []
ctr=0
for ix in x_test:
    if ix == "":
        continue
    ctr+=1
    if ctr%100==0:
        print(ctr)
    path = images_path + ix
    img = image.load_img(path,target_size=(224,224))
    img_array = image.img_to_array(img)
    test_data.append(img_array)
test_data = np.array(test_data)
```

```python
test_data = np.array(test_data)
print(test_data.shape)
```

```python
val_data = []
ctr=0
for ix in x_val:
    if ix == "":
        continue
    ctr+=1
    if ctr%100==0:
        print(ctr)
    path = images_path + ix
    img = image.load_img(path,target_size=(224,224))
    img_array = image.img_to_array(img)
    val_data.append(img_array)
val_data = np.array(val_data)
```

```python
print(train_data.shape,test_data.shape,val_data.shape)
```

```python
model_test = ResNet50(include_top=False,weights='imagenet',input_shape=(224,224,3))
model_test.summary()
```

```python
test_emb = model_test.predict(test_data).reshape((1000,2048))
print(test_emb.shape)
```

```python
train_emb = model_test.predict(train_data).reshape((6000,2048))
print(train_emb.shape)
```

```python
val_emb = model_test.predict(val_data).reshape((1000,2048))
print(val_emb.shape)
```

```python
print(test_emb.shape)
print(train_emb.shape)
print(val_emb.shape)
```

```python
np.save('./test_emb.npy',test_emb)
np.save('./train_emb.npy',train_emb)
np.save('./val_emb.npy',val_emb)
```
